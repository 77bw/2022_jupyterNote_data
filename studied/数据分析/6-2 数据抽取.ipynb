### 一、数据处理
     数据处理时数据价值关键的步骤。数据分析过程中，数据处理占据了很大的工作量，，同时也时整个数据分析中最为重要的环节。
     为什么要数据处理？
     1.海量的原始数据中存在大量不完整，不一致，异常数据，严重影响数据分析结果。
     2.数据处理就是要处理缺失数据以及清除无意义的数据，如删除原始数据集中的无关数据，重复数据，平滑噪声数据，处理缺失值，异常值。

#### 3.记录的抽取：根据一定的条件进行抽取，使用dataframe[condition]抽取,conditon:过滤条件。放回值为一个数据框。
        常用的过滤条件：
            (1) 比较运算：==，>,<,>=,<=,!=
            (2) 范围运算：between(left,right)
            (3) 空值配置：isnull(column)
            (4) 字符匹配：str.contains(patten,na=False)
            (5) 逻辑运算：& | not。

```python
df3[df3.学号==2308024241]
```

```python
df3[df3.学号.between(2308024241,2308024251)]
```

```python
#逻辑条件
df3[(df3.学号>=2308024241) & (df3.学号<=2308024251)]
```

- 按索引条件抽取
    - 雷同DataFrame访问，<br>
（1）抽取某行数据,df.loc[行索引名]<br>
（2）抽取几行数据，df.loc[起始行名：结束行名(含)]，不写列名，表示所有列<br>
（3）抽取列数据：df.loc[:,列名]，或者，df[列名]<br>
（3）df.iloc[],用序号抽取。<br>如：df.iloc[1,0]第2行第1列；df.iloc[0:2,:]第1到2行所有列；df.iloc[[0,2],:]第1和3行所有列；df,iloc[:,1]第2列；df.iloc[2,:]第2行所有列。
- 随机抽样
    - 随机从数据中按照一定的行数或比例抽取数据<br>
np.random.randint(start,end,num)<br>
