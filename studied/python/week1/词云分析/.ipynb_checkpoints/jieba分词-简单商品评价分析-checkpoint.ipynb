```python
import jieba
```

### jieba分词有三种模式
    1.全模式cut，第二个参数cat_all=True
    2.精确模式cut,第二个参数cat_all=Flase 默认
    3.搜索引擎模式：cut_for_search函数

```python
tmpstr = "郭靖和哀牢山三十六剑。"
res = jieba.cut(tmpstr) #精确模式
print(' '.join(res)) #是一个可迭代的generator，可以使用for循环来遍历结果，本质上类似list
```

```python
tmpstr = "郭靖和哀牢上三十六剑。"
res = jieba.cut(tmpstr,cut_all=True) #全模式
print(' '.join(res)) #是一个可迭代的generator，可以使用for循环来遍历结果，本质上类似list
```

```python
res = jieba.cut_for_search(tmpstr) #搜索引擎模式
print(' '.join(res)) #是一个可迭代的generator，可以使用for循环来遍历结果，本质上类似list
```

```python
import pandas as pd
```

```python
file_data=pd.read_csv("comment.csv",encoding="GBK")
file_data
```

```python
file_data=file_data.drop_duplicates()
file_data
```

```python
file_data.info()
```

```python
#3. 分词，jieba分词，精确模式
cut_words=jieba.lcut(str(file_data['评价信息'].values),cut_all=False)
cut_words
```

```python
# 4.加载停用词表，并在评价中去掉停用词
with open('./stoplist.txt') as f:
    #打开停用词文件，读取里面的词
    stop_words=f.read()
    new_data=[]
    for word in cut_words:
        if word not in stop_words:
            new_data.append(word)
new_data
```

```python
# 5.词频统计
#1.导入包
import nltk
from nltk import FreqDist #统计词频
freq_list=FreqDist(new_data)
#取出高频词
most_common_words=freq_list.most_common()
most_common_words
```
