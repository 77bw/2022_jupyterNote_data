```python
#1.爬取第一本书西游记
url="https://www.shicimingju.com/book/xiyouji.html"
res=requests.get(url) #获得整html页面
res.encoding=res.apparent_encoding
html=res.text
html
```

```python
#2.使用BeautifulSoup,将已经下载的html内容解析为soup对象，每个节点就是一个对象
soup=BeautifulSoup(html,'lxml')
soup
```

```python
#4。过滤节点元素，class="book-mulu"
orpath=soup.find_all(class_="book-mulu")
orpath
```

```python
# 5使用正则表达式去除多余的标记，获取四大名著里面所有的章节的超链接：如a href="/book/xiyouji/1.html" 得到/book/xiyouji/1.html

pattern=re.compile('(?<=href=").*?(?=")')
chape_list=re.findall(pattern,str(orpath))
chape_list
```

```python
web="https://www.shicimingju.com/book/xiyouji/1.html"
res=requests.get(web)
res.encoding=res.apparent_encoding
soup=BeautifulSoup(res.text,'lxml')
contents=soup.find_all(class_="card bookmark-list")
contents
```

```python
#6.具体进入每一个章节的超链接，获取章节的内容
url_base="https://www.shicimingju.com"
for l in chape_list:
    web=url_base+l
    res=requests.get(web)
    res.encoding=res.apparent_encoding
    html=res.text
    pattern=re.compile('(?<=<p>).*?(?=</p>)')
    text=re.findall(pattern,str(html))
    print(text)
```
