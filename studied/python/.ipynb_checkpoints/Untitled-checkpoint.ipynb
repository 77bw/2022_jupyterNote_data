1、提取BX-Users数据

```python
import pandas as pd
import numpy as np
names=["User-ID","Location","Age"]
fime=pd.read_csv("BX-Users.csv",skiprows=[0],names=names)
#数据清洗
fime=fime.replace(" ",np.nan)     #替换空格符号
fime=fime.dropna(axis=0,how="any")   #删除含有空值行
#删除国家年龄为null的数据
fime=fime[~ fime['Age'].str.contains('NULL')]
fime

```

```python
#BX-Book-Ratings.csv表数据
fime1=pd.read_csv("BX-Book-Ratings.csv",sep=";")
#fime1

#BX_Books.csv
fime2=pd.read_csv("BX_Books.csv",sep=";")
#fime2

#连接
fime3 = pd.merge(fime1,fime2,how='outer',on='ISBN')
fime3=fime3.dropna(axis=0,how="any")   #删除含有空值行
fime3['User-ID'] = fime3['User-ID'].astype('int')
#fime3

#转换数据类型
fime5=fime
fime5['User-ID']=fime5['User-ID'].str.split(';',expand=True)[0]
fime5=fime5.dropna(axis=0,how="any")   #删除含有空值行
fime5['Age']=fime5['Age'].str.extract('(\d+)', expand=False).astype(float)
fime5['User-ID'] = fime5['User-ID'].astype('int')
#fime5

#连接
fime4 = pd.merge(fime5,fime3,how='outer',on='User-ID')
fime4=fime4.dropna(axis=0,how="any")   #删除含有空值行
fime4



```

```python
#获取国家年龄在66到70的用户数据
fime4[(fime4['Age']>=66) & (fime4['Age']<=70)]
#fime4[fime4['Age']>60]
```
