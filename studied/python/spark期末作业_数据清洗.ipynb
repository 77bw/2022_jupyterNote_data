```python
#1.导入相应包
import pandas as pd
import numpy as np
import codecs
```

## 1.1数据清洗有些无的数据（地点无,国家年龄无）

```python

#导入数据
names=["User-ID","Location","Age"]
data=pd.read_csv(r"C:\Users\bw\Desktop\SPARK期末\BX-Users.csv",skiprows=[0],names=names,encoding = "ISO-8859-1")
#查看每列的信息
data.info()
```

```python
#缺失值处理，去除空白值
data=data.replace(" ",np.nan)
data=data.dropna(axis=0,how="any") 
data
```

```python
#删除国家年龄为null的数据
data=data[~ data['Age'].str.contains('NULL')]
data
```

## 1.2数据条件提取，去除缺失值以及获取相应有关的数据，要求三表合一

```python
#合并三表
data1=pd.read_csv(r"C:\Users\bw\Desktop\SPARK期末\BX-Book-Ratings.csv",sep=";",encoding = "ISO-8859-1")
data2=pd.read_csv(r"C:\Users\bw\Desktop\SPARK期末\BX_Books.csv",sep=";",encoding = "ISO-8859-1")
data3 = pd.merge(data1,data2,how='outer',on='ISBN')

#转换数据类型
data4=data
data4['User-ID']=data4['User-ID'].str.split(';',expand=True)[0]
data4=data4.dropna(axis=0,how="any")   #删除含有空值行
data4['Age']=data4['Age'].str.extract('(\d+)', expand=False).astype(float)
data4['User-ID'] = data4['User-ID'].astype('int')

data_all=pd.merge(data3,data4,how='outer',on='User-ID')
data_all.head()
```

```python
#数据预处理
data_all=data_all.dropna(axis=0,how="any")   #删除含有空值行
data_all['User-ID'] = data_all['User-ID'].astype('int')
data_all.head()
```

```python
#查看数据
data_all['Image-URL-L']#经检查Image-URL-S，Image-URL-M，Image-URL-L获取的都是同一个url，所以去除其他两个
```

```python
#删除数据分析不需要的列，剩下有用的用户，图书等级，标题，作者，出版时间，出版社，url,位置，国家年龄
data_all=data_all.drop(['ISBN','Image-URL-M','Image-URL-L'],axis=1)
data_all.head()
```

## 1.3数据范围抽取，要求定义一个范围值使用3个表后抽取数据显示（获取图书等级为5，国家年龄>50的有关信息）

```python
df=data_all[(data_all['Book-Rating']==5) & (data_all['Age']>50)]
df
```
