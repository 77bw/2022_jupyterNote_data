encode(编码)和decode(解码)的区别:
在运行的时候，有时候会报编码错误，本文就来研究一下这个问题。

为什么会出现乱码呢？因为在文件存的时候格式和读取时候格式不一致就会乱码了。

字符串在python内部的表示是unicode编码，也可以说现在的内存是unicode编码格式，硬盘是utf-8。平常的数据操作都是先把数据读取到内存中，所以内存中都是unicode编码格式。所以我们平常在做编码转换时候，通常用unicode作为中间编码。先将其他编码的字符串解码(decode)成unicode,再从unicode编码(encode)成另一种编码格式。

decode的作用是将二进制数据解码成unicode编码，如str1.decode('utf-8'),表示将utf-8的编码字符串解码成unicode编码。

简单的来说：decode就是把二进制数据(bytes)转化成人看的懂得英文或者汉字(decode用的比较多)

encode的作用是将unicode（统一码）编码的字符串编码成二进制数据，如str2.encode('utf-8'),表示将unicode编码的字符串编码成utf-8。

编码是将字符串转换为字节数组的过程，解码是将字节数组转换成字符串的过程。

```python
编码转换：
数据在网络上传输就必须是bytes方式的数据
encode(编码)和decode(解码)
对方怎么编码就得用什么方式去解码
解决：先提取网页本身的编码，再按照该编码去解码，即可
```

```python
a="你好"
b=a.encode()
print(type(a))
print(b)
print(b.decode())
print(type(b))
```

## urllib库的使用

```python
#urllib爬取
from urllib import request #引入包
url="https://www.bilibili.com"
#设置头部，防止反爬
header={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"}
req=request.Request(url,headers=header) #构建Reuest对象，传入字典形式的headers属性，模拟浏览器。
response=request.urlopen(req) #爬取网页，得到响应。
print(response)
data=response.read().decode() #获取网页后的html文件需要进行解码才能显示。
print(data)
file_path="./bilibli首页1.html"   #创建文件地址，如果有重新写入。
with open(file_path,"w",encoding="utf-8") as f:  #写入文件也要进行解码 W打开文件只用于写入，文件已存在，删除内容重新写入。目由则创建写入。
    f.write(data)
```

1、json.dumps()和json.loads()是json格式处理函数（可以这么理解，json是字符串）
　　(1)json.dumps()函数是将一个Python数据类型列表进行json格式的编码（可以这么理解，json.dumps()函数是将字典转化为字符串）
　　(2)json.loads()函数是将json格式数据转换为字典（可以这么理解，json.loads()函数是将字符串转化为字典）

2、json.dump()和json.load()主要用来读写json文件函数

区分json与dict
json:
本质是一种文件组织方式，比如你熟悉的txt, csv, doc,docx,xls,xlsx文件等等。
dict：
是程序语言里定义的一种数据结构，比如列表list，比如集合set，比如字符串str，有比如数组array。

```python
from selenium import webdriver
import time
path="D:\AppData\chromedriver.exe"  #可以指定驱动的位置，也可以放在.py同级目录下就不用设置path了
browser=webdriver.Chrome(path)
url="https://top.baidu.com/board?platform=pc&sa=pcindex_entry"
browser.get(url) #访问页面
buton=browser.find_element_by_css_selector('#sanRoot > main > div.hot-wrap_1nNog > div.theme-hot.category-item_1fzJW > div.title-wrap_256Xx > a.more_Uf8LD > span')
buton.click()#点击展开 
time.sleep(1)
#print(browser.page_source)  #page_source用户获取网页的完整html
#如果xpath中有""  那么外面就得用单括号''括起来
c=browser.find_elements_by_xpath('//*[@id="sanRoot"]/main/div[2]/div/div[2]/div/div[2]/a/div[1]') #elements带s返回所有结果，目由就返回第一个
for i in c:
    print(i.text)
browser.close()
```
