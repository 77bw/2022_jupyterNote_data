## urllib库的使用

```python
#urllib爬取
from urllib import request #引入包
url="https://www.bilibili.com"
#设置头部，防止反爬
header={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"}
req=request.Request(url,headers=header) #构建Reuest对象，传入字典形式的headers属性，模拟浏览器。
response=request.urlopen(req) #爬取网页，得到响应。
print(response)
data=response.read().decode()# 获取网页后的html文件需要进行解码才能显示。
print(data)
file_path="./bilibli首页1.html"   #创建文件地址，如果有重新写入。
with open(file_path,"w",encoding="utf-8") as f:  #写入文件也要进行解码 W打开文件只用于写入，文件已存在，删除内容重新写入。目由则创建写入。
    f.write(data)
```

## BeautifulSoup的简单使用 
解析网页的内容
select和find_all获取网页内容为list类型（组标签），得到内容应先[0]
select_one和find获取的是标签，内容可以直接得到

```python
import requests
from bs4 import BeautifulSoup
url="https://www.bilibili.com"
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36'}
res=requests.get(url,headers=headers)#爬取网页，得到响应。
res.enconding="utf-8" #进行解码显示

'''
    解析器:
        Python  内置标准库  优势:执行速度适中,文档容错能力强
            BeautifulSoup(html,'html.parser')  3.7 or 3.2 容错能力较差
        lxml HTML   速度快,文档容错能力强  (最常用)
            BeautifulSoup(html,'lxml')
        lxml XML    速度快 ,唯一支持xml解析
            BeautifulSoup(html,'xml')
        html5lib    最好的容错性,速度慢,不依赖外部扩展
            BeautifulSoup(html,'html5lib')
'''
soup=BeautifulSoup(res.text,'lxml')
soup
a=soup.find('title').text  #find只获取title的第一个，text为获得文本
print(a)
```

## 使用re（正则表达式）进行精确数字的匹配


```python
#1.re正则表达式
# .* 的话就可以匹配任意长度的任意字符 贪婪匹配 上面这个只会匹配到剩下最后一个
# .*? 的话就可以匹配任意长度的任意字符 非贪婪匹配  这个只要匹配到数字就停止
```

```python
# 贪婪模式与非贪婪模式
import re
string ='<你爱我我爱你> <蜜雪冰城甜蜜蜜>'
result = re.findall(r"<.*>" ,string)   #贪婪匹配发现>这个被吃掉了，到最后>才返回
result1 = re.findall(r"<.*?>" ,string)  #第一次遇到>就把前面的东西返回
print(result)
print(result1)
```

```python
# 正则表达式  /d 匹配任意数字   .* 的话就可以匹配任意长度的任意字符
#贪婪匹配
import re
s='hello 1234567 world'
res = re.match('he.*(\d+).*7',s)
print(res.group(1))
#在贪婪匹配下，.* 会匹配尽可能多的字符。正则表达式中.*后面是\d+，也就是至少一个数字，并没有指定具体多少个数字，
#因此，.*就尽可能匹配多的字符，这里就把123456匹配了，给\d+留下个可满 足条件的数字 7.最后得到的内容就只有数字7了。
```

```python
#非贪婪模式
import re
s='hello 1234567 world'
res = re.match('he.*?(\d+).*rld$',s)
print(res.group(1))
#此时就可以成功获取1234567了。原因可想而知，贪婪匹配是尽可能匹配多的字符，非贪婪匹配就是尽叮能匹配少的字符。
#当.*?匹配到hello后面的空白字符时，再往后的字符就是数字了，而\d+恰好可以匹配，那么这里. *?就不再进行匹配，
#交给\d+去匹配后面的数字。所以这样，*?匹配了尽可能少的字符，\d+ 的结果就是1234567了。
```

```python
#2.re模块
1）查找模块
search 只返回一个
match 只返回一个，必须从一开始进行匹配，开通匹配不到就没有
finditer 返回match迭代器
（上面三个都返回match对象，调用数据都得用group()获取）
findall 只返回字符串
```

```python
import re
text = '18812345678，他还有一个电话号码是18887654321，他爱好的数字是01234567891，他的座机是∶0571-52157216,她的座机号码是：0163-77778888'
m=re.search(r'(\d{4})-(\d{8})',text)  #r去除转义字符
print(m.group())
```

```python
import re
text = "xaBc, Abc, ABC"
m = re.search(r'abc',text,flags=re.I) #re.I不区分大小写
x = re.match(r'abc',text,flags=re.I)  #xaBc match得从头开始一开始的x不是所有匹配不到
d=re.findall(r'abc',text,flags=re.I)  #findall返回的是列表
it=re.finditer(r'abc',text,flags=re.I)
flags= re.I
print(m.group())
print(x)
print(d)
for i in it:
    print(i)
```

```python
import re
a = "123abc456"
#groups
print(re.search("([0-9]*)([a-z]*)([0-9]*)",a).groups())   #以元组形式返回全部分组截获的字符串
print(re.search("([0-9]*)([a-z]*)([0-9]*)",a).group(0))   #123abc456,返回整体  匹配正则表达式整体结果
print(re.search("([0-9]*)([a-z]*)([0-9]*)",a).group(1))   #匹配第一个待匹配的值123
print(re.search("([0-9]*)([a-z]*)([0-9]*)",a).group(2))   #匹配带匹配值的第二个
print(re.search("([0-9]*)([a-z]*)([0-9]*)",a).group(3))   #匹配带匹配值的第三个
```

## selenium库的使用
获取元素可以使用xpath路径,在控制台验证xpath按下cirl+f
百度的数据页面使用了动态渲染技术，我们可以用selenium来爬取,在爬虫中主要解决JavaScript渲染问题
selenium是一个用于web应用程序测试的工具,直接运行在浏览器中，就像真正的用户在操作一样
当requests库无法正获获取网页的内容的时候，就可以尝试使用selenium来爬取信息
步骤：
1创建浏览器对象②浏览器.get()③浏览器.find0 

```python
from selenium import webdriver
import time
path="D:\Data\jupyterNote Data\chromedriver.exe"  #可以指定驱动的位置，也可以放在.py同级目录下就不用设置path了
browser=webdriver.Chrome(path)
url="https://top.baidu.com/board?platform=pc&sa=pcindex_entry"
browser.get(url) #访问页面
buton=browser.find_element_by_css_selector('#sanRoot > main > div.hot-wrap_1nNog > div.theme-hot.category-item_1fzJW > div.title-wrap_256Xx > a.more_Uf8LD > span')
buton.click()#点击展开 
time.sleep(1)
#print(browser.page_source)  #page_source用户获取网页的完整html
#如果xpath中有""  那么外面就得用单括号''括起来
c=browser.find_elements_by_xpath('//*[@id="sanRoot"]/main/div[2]/div/div[2]/div/div[2]/a/div[1]') #elements带s返回所有结果，目由就返回第一个
for i in c:
    print(i.text)
browser.close()
```
