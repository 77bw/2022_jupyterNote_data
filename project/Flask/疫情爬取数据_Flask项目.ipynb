```python
#3.存储数据 其中history 表存储每日总数据，details表存储每日详细数据
# 安装pip install pymysql
#(1)建立连接
#（2）建立游标
#（3）执行操作
#（4）关闭连接
import pymysql
import traceback #追踪异常
```

```python
#（1）建立连接
def get_conn(): #封装连接
    """
    :return: 连游标
    """
    #创建连接
    conn=pymysql.connect(host="localhost",
                        user="root",
                        password='123456',
                        db="cov1",
                        charset='utf8')
    #创建游标
    cursor=conn.cursor()#执行完毕返回的结果集默认以元组显示
    return conn,cursor
```

```python
#（2）关闭连接
def close_conn(conn,cursor): #执行完毕关闭连接和游标
    conn.close()
    cursor.close()
```

```python
#写入数据库
#爬取百度热搜数据
def get_baidu_hot():
    url = "https://top.baidu.com/board?tab=realtime"
    option = ChromeOptions()
    option.add_argument("--headless")  # 隐藏游览器
    option.add_argument("--no--sandbox")
    browser = Chrome(options=option)
    browser.get(url)  # 2.打开地址
    time.sleep(1)
    # 爬虫与反爬，模拟人等待1秒
    c = browser.find_elements_by_xpath('//*[@id="sanRoot"]/main/div[2]/div/div[2]/div/div[2]/a/div[1]')  #共有30条记录
    context = [i.text for i in c]
    print(context)
    browser.close()
    return context
```

```python
def update_hotsearch():
    cursor = None
    conn = None
    try:
        context = get_baidu_hot()
        print(f"{time.asctime()}开始更新数据")
        conn,cursor = get_conn()
        sql = "insert into hotsearch(dt,content) values(%s,%s)"
        ts = time.strftime("%Y-%m-%d %X")
        for i in context:
            cursor.execute(sql,(ts,i))
        conn.commit()
        print(f"{time.asctime()}数据更新完毕")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn,cursor)

if __name__ == "__main__":
    #insert_history()
    #update_history()
    #update_details()
    update_hotsearch
```
