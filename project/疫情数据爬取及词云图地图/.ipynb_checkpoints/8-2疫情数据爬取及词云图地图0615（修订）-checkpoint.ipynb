```python
#网页内容抓取
#1.导入包
import requests
import xlwt #操作excel
import json
import re
#2.定义一个查询
province=input("请输入查询的省份：")
#3.定位新冠肺炎国内的url,找到json文件。
url="https://api.inews.qq.com/newsqa/v1/query/pubished/daily/list?province="
#4.伪装浏览器
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36'}
#response=requests.get(url)
#5.查询字段
data={"province":province}
#6.发出请求
page_text=requests.post(url=url,headers=headers,data=data).json()
#7.打印json文件
#fileName=province+'.json'
#fp=open(fileName,'w',encoding='utf-8')
#json.dump(page_text,fp=fp,ensure_ascii=False)
#8.提取json文件中的data键
data=page_text['data']
#9.查看提取data键内容
#data
#10.查看总提取的日期天数
dada_num=len(data) #计算总共统计了多少天
book=xlwt.Workbook(encoding='utf-8') # Python中使用的是xlwt模块来生成Excel文件，并且可以控制单元格的格式。
                                     # book = Workbook(encoding=’utf-8’)，就可以在excel中输出中文了。默认是ascii

sheet=book.add_sheet('疫情数据',cell_overwrite_ok=True) # add_sheet(sheet_name): 添加sheet

#列名
col_name=['省份','日期','累计确诊','累计死亡','累计治愈','境外输入']
#11，先将列名写入到表中
for i in range(0,6):
    sheet.write(0,i,col_name[i])
#12.将提取的具体数据写入excel表
for i in range(dada_num):
    sheet.write(i+1,0,province)
    sheet.write(i+1,1,data[i]['date'])
    sheet.write(i+1,2,data[i]['confirm'])
    sheet.write(i+1,3,data[i]['dead'])
    sheet.write(i+1,4,data[i]['heal'])
    
    #提取境外输入的病例，使用正则表达式
    ex='确诊病例(.*?)例 | 新增(.*?)例'
    #配置表达式,使用贪婪模式
    match=re.findall(ex,data[i]['description'])
    
    if len(match)==0:
        sheet.write(i+1,5,0)
    else:
        sheet.write(i+1,5,int(match[0]))
    print("%s数据爬取结束"%(data[i]['date']))
#13.将爬取到的文件存入到excel表中
book.save(province+'疫情数据'+'.xls')
print("数据爬取结束。。。。。")
            
    
    
                              
                              
    
    

```

```python
# 全国省份
province_name = ['北京', '上海', '江苏', '广东', '浙江', '陕西', '山西','河北','河南','湖北','湖南','安徽','山东','重庆','四川','辽宁',
                     '吉林','黑龙江','天津','福建','云南','贵州','青海','海南','西藏','广西','宁夏','甘肃','新疆','内蒙古','江西']

for item in range(len(province_name)):
    
    province = province_name[item]
    
    url = 'https://api.inews.qq.com/newsqa/v1/query/pubished/daily/list?province='
    headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'
        }
    data = {
            'province': province
        }
    page_text = requests.post(url=url, data=data, headers=headers).json()
        # fileName = province+'.json'
        # fp = open(fileName,'w',encoding='utf-8')
        # json.dump(page_text,fp=fp,ensure_ascii=False)
    data = page_text['data']
        # print(data)
    date_num = len(data)  # 计算总共的天数

    book = xlwt.Workbook(encoding='utf-8')
    sheet = book.add_sheet('疫情数据', cell_overwrite_ok=True)
    col_name = ['省份', '日期', '累计确诊', '累计死亡', '累计治愈', '境外输入']
    for i in range(0, 6):
        sheet.write(0, i, col_name[i])

    for i in range(date_num):
        sheet.write(i + 1, 0, province)
        y = data[i]['date'].split('.')[0]
        m = data[i]['date'].split('.')[1]
        sheet.write(i + 1, 1, y+'月'+m+'日')
        sheet.write(i + 1, 2, data[i]['confirm'])
        sheet.write(i + 1, 3, data[i]['dead'])
        sheet.write(i + 1, 4, data[i]['heal'])

            # 提取境外输入病例
        ex = '确诊病例(.*?)例 | 新增(.*?)例'

        match = re.findall(ex, data[i]['description'])
        
        if len(match) == 0:
            sheet.write(i + 1, 5, 0)
        else:
            sheet.write(i + 1, 5, int(match[0]))

    print('%s   爬取结束'%province)

    book.save('./全国各地疫情数据/'+province + '疫情数据' + '.xls')

print('爬取结束----')
```

## 数据提取

```python
import pandas as pd
import xlwt
import numpy as np
from datetime import datetime
date = input('请输入想要提取的日期（格式如08月01日）:')


book = xlwt.Workbook(encoding='utf-8')
sheet = book.add_sheet('同一日期提取',cell_overwrite_ok=True)
col_name = ['省份','累计确诊', '累计死亡', '累计治愈', '境外输入']
for i in range(0,5):
    sheet.write(0,i,col_name[i])

    province_name = ['北京', '上海', '江苏', '广东', '浙江', '陕西', '山西', '河北', '河南', '湖北', '湖南', '安徽', '山东', '重庆', '四川', '辽宁',
                     '吉林', '黑龙江', '天津', '福建', '云南', '贵州', '青海', '海南', '西藏', '广西', '宁夏', '甘肃', '新疆', '内蒙古', '江西']

    for i in range(len(province_name)):
        province = province_name[i]
        # print(province)
        filename = './全国各地疫情数据/'+province+'疫情数据.xls'
        data  = pd.read_excel(filename)
        deal_num = data['日期']
        for j in range(len(deal_num)):
            if deal_num[j] == date:
                sheet.write(i+1,0,province)
                sheet.write(i+1,1,int(data['累计确诊'][j]))
                sheet.write(i+1,2,int(data['累计死亡'][j]))
                sheet.write(i+1,3,int(data['累计治愈'][j]))
                sheet.write(i+1,4,int(data['境外输入'][j]))
        print('%s提取完毕'%province)

    book.save('全国各地'+date+'提取结果.xlsx')
print('处理完毕......')
```

```python
from pyecharts.charts import Pie, Map, Bar, Line, Grid, Page,Funnel #插件可以绘制很酷炫的图形
# 1.读取数据
wb = openpyxl.load_workbook('data.xlsx')
# 2.获取工作表
ws = wb['国内疫情']
#5.我们创建一个字典将他们关联，将省份和确诊人数关联
frequency_in = {}
frequency={}
for row in ws.values:
    #print(row)#3.我们看到所需要的数据以元组形式全打印出来啦
    #4.我们去掉表头第一列标题，通过一个if语句
    if row[0]=='省份':
        pass
    else:
        frequency[row[0]]=float(row[1]) #将省份和确诊人数关联,可以不转换数据类型
#6.打印查看是否写入字典
print(frequency)
provice = list(frequency.keys()) 
num = list(frequency.values()) 
```

```python
### 3.画图
from pyecharts import options as opts
import datetime
date = datetime.datetime.now().year


pieces = [{"max": 999999, "min": 1001, "label": ">10000", "color": "#8A0808"},
                  {"max": 9999, "min": 1000, "label": "1000-9999", "color": "#B40404"},
                  {"max": 999, "min": 100, "label": "100-999", "color": "#DF0101"},
                  {"max": 99, "min": 10, "label": "10-99", "color": "#F78181"},
                  {"max": 9, "min": 1, "label": "1-9", "color": "#F5A9A9"},
                  {"max": 0, "min": 0, "label": "0", "color": "#FFFFFF"},
                  ]

chinaMap =Map(init_opts=opts.InitOpts(width = '1000px', height='880px'))
chinaMap.add("累计确诊人数", [list(z) for z in zip(provice,num)], "china", is_map_symbol_show=False)
chinaMap.set_global_opts(
                title_opts=opts.TitleOpts(title="中国疫情地图分布", subtitle='截止%s中国疫情分布情况'%(date), pos_left="center", pos_top="10px"),
                legend_opts=opts.LegendOpts(is_show=False),
                visualmap_opts=opts.VisualMapOpts(max_=200, is_piecewise=True,
                                                  pieces=pieces,
                                                  ))
               
    
chinaMap.render("中国疫情地图.html")
```
